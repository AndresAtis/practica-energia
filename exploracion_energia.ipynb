# Ejercicio 2

### Convierte la columna de fecha a datetime
### La fecha viene como texto — necesitas convertirla para poder operar con ella
### Como ingeniero: esto es como convertir unidades. El dato existe pero en formato inútil hasta que lo transformas.

```python
df['Datetime'] = pd.to_datetime(df['Datetime']) #Convierte el valor a data time
```

```python
print(df.dtypes)
```

```python
# Se usan para agrupar/extraer fechas 

df['Year'] = df['Datetime'].dt.year
df['Month'] = df['Datetime'].dt.month
df['Hour'] = df['Datetime'].dt.hour
```

```python
print(df[['Datetime', 'Year', 'Month', 'Hour']].head())
```

# Ejercicio 4
## Consumo promedio por hora del día
## ¿A qué hora se consume más energía? Agrupa por hora y calcula el promedio

```python
consumo_por_hora = df.groupby('Hour')['AEP_MW'].mean().round(1) # agrupa por horas los datos de los dias del df y les calcula la media
print(consumo_por_hora)
```

```python
print(consumo_por_hora.sort_values(ascending=False).head()) # organiza los datos de forma descendente y muestra los primeros 5
```

```python
hora_pico = consumo_por_hora.idxmax() # muestra el indice en donde esta ubicado el mayor valor

print(f"Hora de mayor consumo: {hora_pico}:00 hrs")
```

# Ejercicio 5
## Consumo promedio por mes — estacionalidad
## Identifica qué meses tienen mayor demanda energética y por qué
## Como termodinámico: verás dos picos (verano e invierno) por HVAC. El valle en primavera/otoño es cuando la carga térmica es mínima.

```python
consumo_mes = df.groupby('Month') ['AEP_MW'].mean()

meses = ['Ene', 'Feb', 'Mar','Abr' ,'May', 'Jun', 'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']
consumo_mes.index = meses

print(consumo_mes)
```

```python
print(f"\nMes más alto: {consumo_mes.idxmax()}")
print(f"\nMes más bajo: {consumo_mes.idxmin()}")
```

# Ejercicio 6

## Filtra solo los años completos
## Elimina años con datos incompletos para no sesgar el análisis
## Un año completo tiene 8,760 horas. Si un año tiene menos, los datos están incompletos y pueden distorsionar promedios anuales.

```python
registros_por_año = df.groupby('Year')['AEP_MW'].count()
print(registros_por_año)
```

```python
años_completos = registros_por_año[registros_por_año>8000].index
df_clean = df[df['Year'].isin(años_completos)]
```

```python
print(f"\nRegistros originales: {len(df)}")
print(f"\nRegistros limpios: {len(df_clean)}")
df_clean
```

```python
registros_eliminados = len(df)-len(df_clean)
print(registros_eliminados)
```

```python
años_eliminados = registros_por_año[registros_por_año <= 8000]
print(años_eliminados)
```
