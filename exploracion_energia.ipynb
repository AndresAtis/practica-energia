# Ejercicio 2

### Convierte la columna de fecha a datetime
### La fecha viene como texto — necesitas convertirla para poder operar con ella
### Como ingeniero: esto es como convertir unidades. El dato existe pero en formato inútil hasta que lo transformas.

```python
df['Datetime'] = pd.to_datetime(df['Datetime']) #Convierte el valor a data time
```

```python
print(df.dtypes)
```

```python
# Se usan para agrupar/extraer fechas 

df['Year'] = df['Datetime'].dt.year
df['Month'] = df['Datetime'].dt.month
df['Hour'] = df['Datetime'].dt.hour
```

```python
print(df[['Datetime', 'Year', 'Month', 'Hour']].head())
```

# Ejercicio 3

## Calcula media, mínimo, máximo y desviación estándar del consumo energético
## Pregunta: ¿Qué tan variable es el consumo? Compara la desviación estándar con la media — ese ratio te dice mucho.

```python
df
```

```python
print(df['AEP_MW'].describe()) #Da una descripcion estadistica de la columna seleccionada
```

```python
print(f"Media: {df['AEP_MW'].mean():.1f} MW") #Muestra la media ademas :.#f se usa para indicar la cantidad de decimales
```

```python
print(f"Maximo: {df['AEP_MW'].max():.1f} MW") #Muestra el maximo
```

```python
print(f"Minimo: {df["AEP_MW"].min():.1f} MW") #Muestra el minimo
```

```python
print(f"Desv. std {df['AEP_MW'].std():.1f} MW") #Muestra la desviacion estandar
```

```python
std = df['AEP_MW'].std()
media = df['AEP_MW'].mean()
CV = (std/media)*100
print(f"Porcentaje de desviación: {CV:.1f} %") #Muestra el porcentaje de variacion
```

# Ejercicio 4
## Consumo promedio por hora del día
## ¿A qué hora se consume más energía? Agrupa por hora y calcula el promedio

```python
consumo_por_hora = df.groupby('Hour')['AEP_MW'].mean().round(1)
print(consumo_por_hora)
```

```python
print(consumo_por_hora.sort_values(ascending=False).head())
```

```python
hora_pico = consumo_por_hora.idxmax()

print(f"Hora de mayor consumo: {hora_pico}:00 hrs")
```
